# Importing Libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/home/mahima/Documents/salary.csv')
data
```

# Split the data based on independent & dependent variables

```python
X=data.iloc[:,:-1] #experience as independent varaible 
X
```

```python
y=data.iloc[:,1:] # salary as dependent variable 
y
```

# Divide the dataset into train & test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=1/3,random_state=0)
```

# Implement the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train) 
```

# Predicting the result

```python
prediction=model.predict(X_test)
prediction
```

# Visualization using matplot

```python
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train, model.predict(X_train))
plt.title('Experience Vs Salary')
plt.xlabel('Experience in Years')
plt.ylabel('Salary')
plt.show()
```
